### Creating conditions to filter information  

```python
#Creating subet for the dataframe
df = ords_prods_merge [:1000000]
```

```python
df.shape
```

```python
df.head()
```

```python
# Defining the conditions

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the condition
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
missing_prices = df['prices'].isnull().sum()
```

##### Using loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

#### Using the the for-loop function 

```python
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creation of a loop function for days of the week
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Creation of a new column for busiest_day
ords_prods_merge['busiest_day'] = result
```

#### Changing datatype for columns 

```python
ords_prods_merge[['product_id', 'aisle_id', 'department_id', 'order_id', 'user_id', 'orders_day_of_week', 'order_hour_of_day']] = ords_prods_merge[['product_id', 'aisle_id', 'department_id', 'order_id', 'user_id', 'orders_day_of_week', 'order_hour_of_day']].astype(str)
```

```python
ords_prods_merge.dtypes
```

#### Creating new labels for busiest days 

```python
result_2 = []

for value in ords_prods_merge["orders_day_of_week"]:
    if value in ['0', '1']:
        result_2.append("Busiest days")
    elif value in ['3', '4']:
        result_2.append("Least busy")
    else:
        result_2.append("Regularly busy")
```

```python
# Creation of a new column for busiest_days
ords_prods_merge['busiest_days'] = result_2
```

```python
# Frequency for the conditions in days
ords_prods_merge['busiest_days'].value_counts (dropna = False)
```

#### Finding the busiest hours

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in ['10', '11','14', '15', '13', '12', '16','9']:
        result_3.append("Most orders")
    elif value in ['3', '4', '2', '5', '1','0','6','23']:
        result_3.append("Fewest orders")
    else:
        result_3.append("Average orders")
```

```python
result_3
```

```python
# Creation of a new column for busiest_period_of_day
ords_prods_merge['busiest_period_of_day'] = result_3
```

```python
# Frequency for the conditions in hours
ords_prods_merge['busiest_period_of_day'].value_counts (dropna = False)
```
