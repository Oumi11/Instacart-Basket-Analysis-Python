#### 2. Dropping a column in the products dataframe

```python
# deleting 1st column prods
df_prods = df_prods.drop(columns=['Unnamed: 0'])
```

```python
df_prods
```

```python
df_orders_products_combined
```

### /!\ As the data set contains the columns _merge, I will delete it to be able to have a new merged data set with a new _merge column for only these 2 specific datasets

```python
# deleting _merge column from df_orders_products_combined
df_orders_products_combined = df_orders_products_combined.drop(columns=['_merge'])
```

```python
df_orders_products_combined.shape
```

### 3. Combining df_prods and orders_products_combined

```python
df_ords_prods_merge = df_orders_products_combined.merge(df_prods, on = 'product_id', indicator = True)
```
